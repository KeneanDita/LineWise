```python
import pandas as pd
import numpy as np
```

```python
teams = pd.read_csv('Datasets/teams.csv')
```

```python
teams
```

```python
X = teams[['athletes', 'prev_medals']].copy()
y = teams[['medals']].copy()
```

```python
X
```

```python
X.shape
```

```python
X['intercept'] = 1
```

```python
X[['intercept', 'athletes', 'prev_medals']]
```

```python
X_T = X.T
```

```python
X_T
```

```python
B = np.linalg.inv(X_T @ X) @ X_T @ y
```

```python
B
```

```python
B.index = X.columns
```

```python
B
```

```python
Predictions = X @ B
```

```python
Predictions
```

```python
SSR = ((y - Predictions) ** 2).sum()
```

```python
SST = ((y - y.mean()) ** 2).sum()
```
